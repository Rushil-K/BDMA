### **Example SQL Queries for Data Analysis**

#### **1\. Identify Machines with Frequent Downtime**

sql  
CopyEdit  
`SELECT machine_id, COUNT(*) AS downtime_count`  
`FROM downtime_logs`  
`WHERE downtime_duration > 10`  
`GROUP BY machine_id`  
`ORDER BY downtime_count DESC`  
`LIMIT 5;`

ðŸ‘‰ *Finds the top 5 machines with the most downtime incidents.*

#### **2\. Detect High Defect Rate Production Shifts**

sql  
CopyEdit  
`SELECT shift, COUNT(*) AS defect_count`  
`FROM production_quality`  
`WHERE defect_flag = 1`  
`GROUP BY shift`  
`ORDER BY defect_count DESC;`

ðŸ‘‰ *Analyzes which shifts produce the most defects.*

#### **3\. Optimize Energy Consumption**

sql  
CopyEdit  
`SELECT machine_id, AVG(energy_usage) AS avg_energy`  
`FROM energy_logs`  
`GROUP BY machine_id`  
`ORDER BY avg_energy DESC`  
`LIMIT 5;`

ðŸ‘‰ *Finds the top 5 machines consuming the most energy.*

#### **4\. Monitor Raw Material Delays**

sql  
CopyEdit  
`SELECT material_id, AVG(delay_time) AS avg_delay`  
`FROM supply_chain_logs`  
`GROUP BY material_id`  
`ORDER BY avg_delay DESC;`

ðŸ‘‰ *Identifies raw materials with the longest delivery delays.*

#### **5\. Improve Worker Productivity**

sql  
CopyEdit  
`SELECT worker_id, AVG(units_produced) AS avg_production`  
`FROM worker_performance`  
`GROUP BY worker_id`  
`ORDER BY avg_production DESC;`

ðŸ‘‰ *Finds the most productive workers to optimize training programs.*
